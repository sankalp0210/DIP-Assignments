```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = cv2.imread('./res/land.png', 0)
out = np.fft.fftshift(np.fft.fft2(x))

arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(np.log(1 + np.abs(out)), 'gray')
arr[0].set_title("FFT of image")

out[:90,125:135] = 0
out[110:,125:135] = 0
out[90:100,:120] = 0
out[90:100,140:] = 0
out[44:50,:] = 0
out[:,76:82] = 0
out[:,178:184] = 0
out[146:152,:] = 0
arr[1].imshow(np.log(1 + np.abs(out)), 'gray')
arr[1].set_title("FFT after making values 0")
plt.show()
```

```python
out = np.abs(np.fft.ifft2(np.fft.ifftshift(out)))

arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(x, 'gray')
arr[0].set_title("Image")
arr[1].imshow(out, 'gray')
arr[1].set_title("FFT")
plt.show()
```

## Explanation :
> Removing the high frequency noise that is present as white lines in the frequency domain.
