```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sample(img, nx, ny):
    res = img[:,::nx]
    res = res[::ny]
    res = np.repeat(res, ny, axis=1)
    res = np.repeat(res, nx, axis=0)
    return res
```

```python
img = cv2.imread('./res/bricks.jpg', 0)
plt.imshow(img, 'gray')
plt.show()
```

```python
out1 = sample(img, 2, 2)
out1 = np.asarray(out1, dtype = complex)

out2 = sample(img, 3, 3)
out2 = np.asarray(out2, dtype = complex)

arr = plt.subplots(3, 2,figsize =(15, 15))[1]
arr[0][0].imshow(np.abs(img), 'gray')
arr[0][0].set_title("Original Image")
arr[0][1].imshow(np.log(1 + np.abs(np.fft.fftshift(np.fft.fft2(img)))), 'gray')

arr[1][0].imshow(np.abs(out1), 'gray')
arr[1][0].set_title("Image with nx = 0.5, ny = 0.5")
arr[1][1].imshow(np.log(1 + np.abs(np.fft.fftshift(np.fft.fft2(out1)))), 'gray')

arr[2][0].imshow(np.abs(out2), 'gray')
arr[2][0].set_title("Image with nx = 0.33, ny = 0.33")
arr[2][1].imshow(np.log(1 + np.abs(np.fft.fftshift(np.fft.fft2(out2)))), 'gray')

plt.show()
```

## Observations:
> When nx=0.5 and ny=0.5, we see that there is no information loss in the frequency domain, but when nx=0.33 and ny=0.33, we can see loss of information. Hence, the nyquist frequency is nx = 0.5 and ny = 0.5
