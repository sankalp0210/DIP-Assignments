```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./res/lena.jpg', 0)
filt = np.array([[0,1,0], [1,2,1], [0,1,0]])/6
f1 = np.fft.fftshift(np.fft.fft2(img))
f2 = np.fft.fftshift(np.fft.fft2(filt, [img.shape[0], img.shape[1]]))

arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(np.abs(np.log(1 + np.abs(f1))), 'gray')
arr[0].set_title("Fourier transform of image")
arr[1].imshow((np.log(1 + np.abs(f2))), 'gray')
arr[1].set_title("Equivalent filter in frequency domain")
plt.show()
```

```python
res = np.multiply(f1, f2)
out = np.abs(np.fft.ifft2(np.fft.ifftshift(res)))
arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(img, 'gray')
arr[0].set_title("Original Image")
arr[1].imshow(out, 'gray')
arr[1].set_title("After Applying filter")
plt.show()
```
