```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./res/rectangle.jpg', 0)

arr = plt.subplots(2,2,figsize =(15,15))[1]
arr[0][0].imshow(img, 'gray')
arr[0][0].set_title("Original Image")
arr[0][1].imshow(np.abs(np.fft.fftshift(np.fft.fft2(img))), 'gray')
arr[0][1].set_title("FFT")

img = np.roll(img, 40, axis=0)
img = np.roll(img, 40, axis=1)

arr[1][0].imshow(img, 'gray')
arr[1][0].set_title("Image after translating")
arr[1][1].imshow(np.abs(np.fft.fftshift(np.fft.fft2(img))), 'gray')
arr[1][1].set_title("FFT of translated")

plt.show()
```

```python
def rotateImage(image, angle):
    image_center = tuple(np.array(image.shape[1::-1]) / 2)
    rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
    return cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
```

```python
img = cv2.imread('./res/rectangle.jpg', 0)
img = rotateImage(img, 45)

arr = plt.subplots(2,2,figsize =(15,15))[1]
arr[0][0].imshow(img, 'gray')
arr[0][0].set_title("Image after rotatong by 45 degrees")
arr[0][1].imshow(np.abs(np.fft.fftshift(np.fft.fft2(img))), 'gray')
arr[0][1].set_title("FFT")

img = rotateImage(img, 45)

arr[1][0].imshow(img, 'gray')
arr[1][0].set_title("Image after rotatong by 90 degrees")
arr[1][1].imshow(np.abs(np.fft.fftshift(np.fft.fft2(img))), 'gray')
arr[1][1].set_title("FFT")

plt.show()
```
