```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipynb.fs.full.funct import applyFilt, im2col, showImages
```

## Using Sobel y filter :-
>Sobel y is a filter that detects horizontal edges and that is what we wanted for box.png.

```python
img = cv2.imread('./res/box.png')

filt = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
out = np.uint8(applyFilt(img, filt))
print(np.sum(out))
showImages(img, out)
```

```python
img = cv2.imread('./res/blur.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
out1 = np.uint8(np.clip(applyFilt(img, filt),0,255))
out2 = np.uint8(np.clip(applyFilt(img, filt.T),0,255))
arr = plt.subplots(1, 3, figsize =(15,15))[1]
arr[0].imshow(img)
arr[0].set_title('Original Image')
arr[1].imshow(out1, cmap = 'gray')
arr[1].set_title('using above filter')
arr[2].imshow(out2, cmap = 'gray')
arr[2].set_title('transpose of above filter')

plt.show()
```

- The transpose of sobel y, becomes sobel x, which detects vertical edges, hence we can see the same in second output.
