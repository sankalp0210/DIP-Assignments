```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipynb.fs.defs.q5 import applyMed
```

### Applied Median filter :-
> The noise perceived in the image is salt and pepper noise, i.e., while applying some filter we have to take a value in middle to prevent getting either of the extremes.
> Hence applying the median filter to remove noise. 


```python
img = cv2.imread('./res/Degraded.jpg')
orig = cv2.imread('./res/Clear.jpg')
k = 5
median = applyMed(img,k)
arr = plt.subplots(1,3, figsize =(15,10))[1]
arr[0].imshow(orig)
arr[0].set_title('Clean Image')
arr[1].imshow(img)
arr[1].set_title('Degraded Image')
arr[2].imshow(median)
arr[2].set_title('Image after cleaning')
plt.show()
```
