```python
import cv2
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readAndConvert(str):
    img = cv2.imread(str,cv2.IMREAD_UNCHANGED)
    return np.int64(np.array(img))
```

```python
def BitQuantizeImage(im, k):
    im = im>>(8-k)
    im = im *(round(255/((1<<k)-1)))
    return np.uint8(im)
```

```python
im = readAndConvert('res/quantize.jpg')

arr = plt.subplots(2,4,figsize =(15,8))[1]
arr[0][0].imshow(im, cmap = 'gray')
arr[0][0].set_title("Original (8 bit)")

bitQuant= []
for i in range(1,8):
    x = math.floor(i/2)
    bitQuant.append(BitQuantizeImage(im,i))
    arr[i%2][x].imshow(bitQuant[-1],cmap = 'gray')
    arr[i%2][x].set_title("Bit Quantized to " + str(i) +  " bit")

bitQuant.append(im)

plt.show()
```

```python
def getBitPlane(im,k):
    res = im << (8-k)
    res = np.uint8(res)
    res = res >> 7
    res = res << (k-1)
    return res
```

```python
im2 = readAndConvert('res/cameraman.png')

arr2 = plt.subplots(2,4,figsize =(15,8))[1]
bitPlane = []
for i in range(1,9):
    x = math.floor((i-1)/2)
    bp = getBitPlane(im2,i)*(255/(1<<i - 1))
    bp = np.uint8(np.round(bp))
    bitPlane.append(bp)
    arr2[(i-1)%2][x].imshow(bitPlane[-1], cmap = 'gray')
    arr2[(i-1)%2][x].set_title("Bit plane " + str(i))

plt.show()
```

```python
def match(im, bitQuant, bitPlane):
    mn = 1<<32
    ans = ""
    val = bitQuant[0]
    for i in range(len(bitQuant)):
        df = np.linalg.norm(bitQuant[i] - im)
        if df < mn:
            val = bitQuant[i]
            mn = df
            ans = "Bit Quantized to " + str(i+1) + " bits"

    for i in range(len(bitPlane)):
        df = np.linalg.norm(bitPlane[i] - im)
        if df < mn:
            val = bitPlane[i]
            mn = df
            ans = "Bit Plane " + str(i+1)

    print(ans)
    return val
```

```python
im = readAndConvert('res/lena.jpg')
bitQuant= []
for i in range(1,8):
    x = math.floor(i/2)
    bitQuant.append(BitQuantizeImage(im,i))
bitQuant.append(im)
```

```python
bitPlane = []
for i in range(1,9):
    bp = getBitPlane(im,i)*(255/(1<<i - 1))
    bp = np.uint8(np.round(bp))
    bitPlane.append(bp)
```

## Lena1

```python
img = readAndConvert('res/lena1.jpg')
out = match(img, bitQuant, bitPlane)

arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(img, cmap = 'gray')
arr[0].set_title("Original")
arr[1].imshow(out, cmap = 'gray')
arr[1].set_title("Matched")
```

## Lena2

```python
img = readAndConvert('res/lena2.jpg')
out = match(img, bitQuant, bitPlane)

arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(img, cmap = 'gray')
arr[0].set_title("Original")
arr[1].imshow(out, cmap = 'gray')
arr[1].set_title("Matched")
```
