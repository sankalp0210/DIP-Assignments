```python
import cv2
from copy import deepcopy
import numpy as np
import matplotlib.pyplot as plt
```

### Function that reads an image and converts it to a numpy array

```python
def readAndConvert(str):
    img = cv2.imread(str)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    return np.int64(np.array(img))
```

### mostFreqCol
Function to get the most frequent colour from the image

```python
def mostFreqCol(a):
    colors, count = np.unique(a.reshape(-1,a.shape[-1]), axis=0, return_counts=True)
    return np.asarray(colors[count.argmax()])
```

### mergeImage
 - Function that takes images as input and replaces the most occuring color in the foreground image with the background image.
 - To improve the results, we can also check cosine similarity between the color of each pixel with the most frequent color and replace it as well if it is greater than a threshold.

```python
def mergeImage(fg, bg):
    out = deepcopy(fg)
    mostFr = mostFreqCol(fg)
    ind = np.where(np.dot(fg,mostFr) > 0.9*(np.linalg.norm(fg,axis=2)*np.linalg.norm(mostFr)))
    out[ind] = bg[ind]
    return out
```

```python
def showIm(im1, im2, out):
    arr = plt.subplots(1,3,figsize =(15,15))[1]
    arr[0].imshow(im1)
    arr[0].set_title("Foreground image")
    arr[1].imshow(im2) 
    arr[1].set_title("Background image")
    arr[2].imshow(out) 
    arr[2].set_title("Result")
    plt.show()
```

```python
fg = readAndConvert('res/fg.jpg')
bg = readAndConvert('res/bg.jpg')
out = mergeImage(fg, bg)
showIm(fg, bg, out)
```
