```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipynb.fs.full.q6 import readAndConvert, histEqualization, histMatching
```

## 1. Applying Histogram Equalization twice

```python
img = readAndConvert('res/eye.png')
out1 = histEqualization(img)
out2 = histEqualization(out1)

arr = plt.subplots(1,3,figsize =(15,15))[1]

plt.gray()
arr[0].imshow(img)
arr[0].set_title("Original")
arr[1].imshow(out1)
arr[1].set_title("Histogram Equalized on input image")
arr[2].imshow(out2)
arr[2].set_title("Histogram Equalized on 1st output")
plt.show()
```

### No Change:
> As we have already applied histogram equalization once, there will not be any change on applying the same transformation further as the histogram has already been equalized.

## 2.1 Similar histograms

```python
im1 = readAndConvert('res/eq1.jpeg')
im2 = readAndConvert('res/eq2.jpg')
out1 = histMatching(im2, im1)

out = histMatching(out1, im1)

arr = plt.subplots(1,3,figsize =(15,15))[1]

plt.gray()
arr[0].imshow(im1)
arr[0].set_title("Original")
arr[1].imshow(out1)
arr[1].set_title("Specified")
arr[2].imshow(out)
arr[2].set_title("Histogram Equalized")
plt.show()
```

### Explanation : 
> As both the images have similar histograms, there is no change to the image after applying histogram equalization.

## 2.2 Dark -> Light Image

```python
im1 = readAndConvert('res/light.jpeg')
im2 = readAndConvert('res/dark.jpg')
out = histMatching(im2, im1)

arr = plt.subplots(1,3,figsize =(15,15))[1]

plt.gray()
arr[0].imshow(im2)
arr[0].set_title("Original")
arr[1].imshow(im1)
arr[1].set_title("Specified")
arr[2].imshow(out)
arr[2].set_title("Histogram Equalized")
plt.show()
```

### Explanation :
> As the specified image is light, so after applying histogram matching, the output image is also light.
