```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def readAndConvert(str):
    img = cv2.imread(str, cv2.IMREAD_GRAYSCALE)
    return np.int64(np.array(img))
```

## 1. When Most Significant Bit is set to 0  - 

```python
img = readAndConvert('res/lena.jpg')

out = img << 1
out = np.uint8(out)
out = out >> 1

arr = plt.subplots(1,2,figsize =(15,7))[1]
bins = [i for i in range(256)]
plt.gray()
arr[0].hist(img.flatten() , bins, range=[0, 255])
arr[0].set_title("Original")
arr[1].hist(out.flatten() , bins, range=[0, 255])
arr[1].set_title("MSB = 0")
plt.show()
```

### Explanation : 
> When most significant bit is set to 0 the range is reduced to half -> [0,127] in this case, the second half of the intensity spectrum gets mapped to the first half of it and hence we can see the shifting in the histogram, i.e., [128, 255] gets mapped to [0, 127]

## 2. When Least Significant Bit is set to 0  - 

```python
img = readAndConvert('res/lena.jpg')

out = img >> 1
out = out << 1

arr = plt.subplots(1,2,figsize =(15,7))[1]

plt.gray()
arr[0].hist(img.flatten() , bins, range=[0, 255])
arr[0].set_title("Original")
arr[1].hist(out.flatten() , bins, range=[0, 255])
arr[1].set_title("LSB = 0")
plt.show()
```

### Explanation : 
> When least significant bit is set to 0, we can see that the odd intensities have become 0 and they have got mapped to the previous intensity.

## 3. Baud Link

- For each pixel, we have 1 byte of information, 1 start bit and 1 stop bit. So, corresponding to each pixel we have 10 bits of data. So for a 512x512 image, total data = 512x512x10.
- time = (data to be transmitted)/speed
- When we have a 56K baud link -> time = (512x512x10)/(56x1000) = 46.811 seconds
- When we have a 3000K baud link -> time = (512x512x10)/(3000x1000) = 0.8738 seconds
