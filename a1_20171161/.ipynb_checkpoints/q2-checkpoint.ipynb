```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
def readAndConvert(str):
    img = cv2.imread(str)
    return np.int64(np.array(img))
```

## Most Frequent Colors
- Function that returns k most occuring colors sorted in decreasing order of count

```python
def mostFreqCol(a,k):
    b = a.reshape(-1,a.shape[-1])
    colors, count = np.unique(b, axis=0, return_counts=True)
    ind = count.argsort()
    colors = colors[ind[::-1]] 
    return np.asarray(colors[0:k])
```

## Linear Contrast Stretching
 - takes image and 2 numbers as input and stretches the image to the range (a,b).
 - the color having the least intensity is mapped to a and the one with the most intensity is mapped to b

```python
def linContrastStretching(im,a,b):
    d = np.max(im)
    c = np.min(im)
    print(c, d)
    x = (b-a)/(d-c)
    im = a + (im - c) * x
    return np.uint8(np.round(im))
```

## Image 1

```python
im = readAndConvert('res/cont1.jpeg')

cont = linContrastStretching(im,50,200)
k = 10

arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(im)
arr[0].set_title("Original")
arr[1].imshow(cont)
arr[1].set_title("Contrast Stretched")

plt.figure()

arr2 = plt.subplots(1,2,figsize =(16,16))[1]
arr2[0].imshow(mostFreqCol(im,k).T, cmap = 'gray', vmin=0, vmax=255)
arr2[1].imshow(mostFreqCol(cont,k).T, cmap = 'gray', vmin=0, vmax=255)

plt.show()
```

### Explanation - 
 - The original image is more towards brighter side, so it has most frequent colors of bright intensities
 - When the image is transformed to (50, 200), the grayish intensities are most frequent.

## Image 2

```python
im = readAndConvert('res/cont2.jpeg')

cont = linContrastStretching(im,50,200)
k = 10

arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(im)
arr[0].set_title("Original")
arr[1].imshow(cont)
arr[1].set_title("Contrast Stretched")

plt.figure()

arr2 = plt.subplots(1,2,figsize =(16,16))[1]
arr2[0].imshow(mostFreqCol(im,k).T, cmap = 'gray', vmin=0, vmax=255)
arr2[1].imshow(mostFreqCol(cont,k).T, cmap = 'gray', vmin=0, vmax=255)

plt.show()
```

### Explanation - 
 - The original image is more towards darker side, so it has most frequent colors of darker intensities that is around 50.
 - When the image is transformed to (50, 200), we have dark as well as bright intensities that are most frequent.

## Difference in the two : 
 - In the first image, we had intensities in the range(158, 254) whereas in second image the intensities were in the range(52, 141).
 - So, when both of them were brought into the same range(50, 200), variaton in first image was more as compared to second.
 - Hence, we perceive more contrast change in first image than second.
