```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipynb.fs.full.q3 import BitQuantizeImage
```

```python
def readAndConvert(str):
    img = cv2.imread(str,cv2.IMREAD_GRAYSCALE)
    return np.int64(np.array(img))
```

```python
def negative(im, k):
    res = 255 - im
    mx = np.max(res)
    res = res*(k/mx)
    return np.uint8(res)
```

```python
im = readAndConvert('res/lena.jpg')

arr = plt.subplots(2,4,figsize =(15,8))[1]

k = 150
neg = negative(im, k)
for i in range(1,9):
    out = BitQuantizeImage(im,i)
    arr[(i-1)%2][(i-1) // 2].imshow(negative(out, k),cmap = 'gray',vmin =0, vmax = 255)
    arr[(i-1)%2][(i-1) // 2].set_title("Bit Quantized to " + str(i) +  " bit")

plt.show()
```

```python
def gammaTransform(im,gamma):
    res = im / 255
    res = np.power(res,gamma)
    res = res * 255
    return np.uint8(res)
```

```python
img = readAndConvert('res/gamma-corr.png')

arr = plt.subplots(2,2,figsize =(15,12))[1]
arr[0][0].imshow(img, cmap = 'gray')
arr[0][0].set_title("Original")

neg = gammaTransform(img, 0.25)

arr[0][1].imshow(neg,cmap = 'gray')
arr[0][1].set_title("Gamma = 0.25")

neg = gammaTransform(img, 0.5)

arr[1][0].imshow(neg,cmap = 'gray')
arr[1][0].set_title("Gamma = 0.5")

neg = gammaTransform(img, 2)

arr[1][1].imshow(neg,cmap = 'gray')
arr[1][1].set_title("Gamma = 2")

plt.show()
```

## Observations
- Gamma correction is a non-linear operation.
- When gamma > 1, a high range of output spectrum is provided to small range of brighter values in input image, hence the darker values in the image gets increased and the image looks darker.
- When gamma < 1, the reverse happens and hence, the image looks brighter.

```python
def piecewiseTransform(im,k1,k2,a,b):
    a = round(a*255)
    b = round(b*255)
    res = np.where((im>a)&(im<=b), k1*im+k2,im)
    res = np.clip(res,0,255)
    
    return np.uint8(res)
```

```python
img = readAndConvert('res/lena.jpg')

arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(img, cmap = 'gray')
arr[0].set_title("Original")

pcw1 = piecewiseTransform(img,0,0,0,0.3)
pcw1 = piecewiseTransform(pcw1,0,0,0.8,1)

pcw2 = piecewiseTransform(pcw1,0,0,0.6,0.8)
pcw2 = piecewiseTransform(pcw2,(4/3),0,0.3,0.6)

pcw3 = piecewiseTransform(pcw1,0,0,0.3,0.6)
pcw3 = piecewiseTransform(pcw3,-2,round(2*255),0.6,0.8)

pcw = np.maximum(pcw2, pcw3)
arr[1].imshow(pcw, cmap = 'gray')
arr[1].set_title("Piecewise Transformed for 1st graph")

plt.show()
```

```python
img = readAndConvert('res/lena.jpg')

arr = plt.subplots(1,2,figsize =(15,15))[1]
arr[0].imshow(img, cmap = 'gray')
arr[0].set_title("Original")

pcw1 = piecewiseTransform(img,0,0,0,0.2)
pcw2 = piecewiseTransform(pcw1,0,round(0.2*255),0.2,0.4)
pcw3 = piecewiseTransform(pcw2,0,round(0.4*255),0.4,0.6)
pcw4 = piecewiseTransform(pcw3,0,round(0.6*255),0.6,0.8)
pcw5 = piecewiseTransform(pcw4,0,round(0.8*255),0.8,1)

arr[1].imshow(pcw5, cmap = 'gray')
arr[1].set_title("Piecewise Transformed for 2nd graph")

plt.show()
```
